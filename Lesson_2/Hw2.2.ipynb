```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '2017_jun_final.csv'
file = pd.read_csv(file_path)
print(file)
```

```python
print(file.head())
```

```python
shape = file.shape
print(f"Кількість рядків: {shape[0]}")
print(f"Кількість стовпців: {shape[1]}")
```

```python
column_types = file.dtypes
print(f'Типи стовпців:\n,{column_types}')
```

```python
missing_values = file.isnull().sum()
total_values = file.shape[0]
missing_percentage = (missing_values / total_values) * 100
formatted_percentage = missing_percentage.round(2).astype(str) + "%"
print(f'Частка пропусків у кожній колонці:\n{formatted_percentage[1:]}')
```

```python
columns_to_drop = file.columns[file.isna().any() & (file.columns != 'Язык.программирования')]
data = file.drop(columns=columns_to_drop)
print(data)
```

```python
missing_values = data.isnull().sum()
total_values = data.shape[0]
missing_percentage = (missing_values / total_values) * 100
formatted_percentage = missing_percentage.round(2).astype(str) + "%"
print(f'Частка пропусків у кожній колонці:\n{formatted_percentage[1:]}')
```

```python
data = data.dropna()
print(data)
```

```python
shape = data.shape
print(f"Кількість рядків: {shape[0]}")
print(f"Кількість стовпців: {shape[1]}")
```

```python
python_data = file[file['Язык.программирования'] == 'Python']
print(python_data)
```

```python
shape = python_data.shape
print(f"Кількість рядків: {shape[0]}")
print(f"Кількість стовпців: {shape[1]}")
```

```python
cities = python_data['Город'].unique()
average_salary_by_city = python_data.groupby('Город')['Зарплата.в.месяц'].mean()
plt.figure(figsize=(12, 6))
plt.bar(cities, average_salary_by_city, color='skyblue')
plt.title('Середня зарплата по містах', fontsize=16, fontweight=800)
plt.xlabel('Місто', fontsize=12)
plt.ylabel('Середня зарплата (на місяць)', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
grouped_data = python_data.groupby('Должность')
print(grouped_data.size())
```

```python
positions = grouped_data.size().index
counts = grouped_data.size().values
explode = (0.1, 0, 0, 0, 0)
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=positions, autopct='%1.1f%%', startangle=90, explode=explode)
plt.title('Частка кожної посади')
plt.show()
```

```python
aggregated_data = grouped_data['Зарплата.в.месяц'].agg(['min', 'max'])
print(aggregated_data)
```

```python
def fill_avg_salary(row):
    return python_data[python_data['Должность'] == row.name]['Зарплата.в.месяц'].mean().round(2)


aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)
print(aggregated_data)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
aggregated_data.plot(kind='bar', y=['min', 'max', 'avg'], ax=ax, rot=0)
ax.set_ylabel('Заробітна плата', fontsize=12)
ax.set_xlabel('Посада', fontsize=12)
ax.set_title('Мінімальне, максимальне та середнє значення для кожної посади', fontsize=16, fontweight=800)
plt.show()
```

```python
aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)
print(aggregated_data['avg'].describe())
```

```python
aggregated_data.to_csv('salary.csv')
```
