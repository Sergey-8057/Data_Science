1. Функція гіпотези лінійної регресії у векторному вигляді

```python
def hypothesis(X, w):
    return np.dot(X, w)
```

2. Функція для обчислення функції втрат у векторному вигляді

```python
def loss_function(X, y, w):
    return np.square(hypothesis(X, w) - y).sum() / (2 * len(X))
```

3. Один крок градієнтного спуску

```python
def gradient_descent_step(X, y, w, learning_rate):
    predictions = hypothesis(X, w)
    gradient = np.dot(X.T, (predictions - y)) / len(y)
    w = w - learning_rate * gradient
    return w
```

> Блок с отступами



```python
# Вибір ознак та цільової змінної
selected_features = ['area', 'bedrooms', 'bathrooms']
target = 'price'

# Створення матриці ознак та вектора реальних значень
X = np.c_[np.ones(len(df)), df[selected_features].values]
y = df[target].values

# Початкові параметри моделі
initial_w = np.zeros(X.shape[1])

# Параметри навчання
learning_rate = 0.00000001
num_iterations = 10000

# Градієнтний спуск
for _ in range(num_iterations):
    initial_w = gradient_descent_step(X, y, initial_w, learning_rate)

# Отримання найкращих параметрів
best_parameters = initial_w
print("Найкращі параметри w:", best_parameters)
```

5. Отримання найкращих параметрів для датасету за допомогою аналітичного рішення

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


# Створення матриці ознак та вектора реальних значень
X_2 = df[selected_features].values
y_2 = df[target].values

# Поділ даних на навчальний та тестовий набір
X_train, X_test, y_train, y_test = train_test_split(X_2, y_2, test_size=0.2,
                                                    random_state=12)

# Створення об'єкта LinearRegression та навчання моделі
model = LinearRegression()
model.fit(X_train, y_train)

# Отримання параметрів моделі
coefficients = np.append(model.intercept_, model.coef_)
print("Найкращі параметри w (LinearRegression):", coefficients)

# Передбачення на тестовом наборі
y_pred = model.predict(X_test)

# Оцінка якості моделі
mse = mean_squared_error(y_test, y_pred)
print("Середньоквадратична помилка (MSE):", mse)
```
